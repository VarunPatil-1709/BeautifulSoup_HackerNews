### Web Scraping Using BeautifulSoup

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

# Initialize empty lists
News_ = []
links_ = []
votes_ = []

# Initial URL
url = "https://news.ycombinator.com/"
url1 = url
conti = True

while conti:
    # Fetch the HTML content
    response = requests.get(url1)
    if response.status_code != 200:
        print(f"Failed to retrieve data. Status code: {response.status_code}")
    else:
        print("Data retrieved successfully.")
    html_doc = response.text
    soup = BeautifulSoup(html_doc, "html.parser")


    titleline_spans = soup.find_all('span', class_='titleline')
    for title_span in titleline_spans:
        news_title = title_span.find("a").get_text()
        news_link = title_span.find("a").get("href")
        News_.append(news_title)
        links_.append(news_link)

    # Find all scores
    score_spans = soup.find_all('span', class_='score')
    for score_span in score_spans:
        score_text = score_span.get_text().strip().split()[0]
        votes_.append(int(score_text) if score_text.isdigit() else 0)

    # Find the "More" link
    more = soup.find("a", class_="morelink")
    if more:
        url1 = url + more['href']
    else:
        print("No more pages found.")
        conti = False

# Create a DataFrame
df = pd.DataFrame({
    'News': News_,
    'Links': links_,
    'Votes': votes_
})

# Display the DataFrame
print(df)

```
